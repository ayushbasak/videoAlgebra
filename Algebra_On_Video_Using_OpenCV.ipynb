```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
#We are going to see the effects of combining two image frames together
```

```python
def capture():
    cap = cv2.VideoCapture(0);
    _,frame = cap.read()
    cap.release()
    
    return frame
```

```python
def perform(arg=None,time_space=0.1):
    try:
        frame01 = capture()
        time.sleep(time_space)
        frame02 = capture()

        if(arg == "ADD"):
            frameSum = frame01 + frame02
        elif(arg == "SUB"):
            frameSum = (frame02 - frame01)
        elif(arg == "MUL"):
            frameSum = frame01 * frame02
        else:
            pass
        frameGray = cv2.cvtColor(frameSum,cv2.COLOR_BGR2GRAY)
        plt.figure()
        plt.imshow(frame01,cmap = plt.cm.binary)
        plt.figure()
        plt.imshow(frame02,cmap = plt.cm.binary)
        plt.figure()
        plt.imshow(frameGray,cmap = plt.cm.binary)
    except:
        print("WRONG Arguments")
        pass
```

```python
# ADDITION of Two Frames
perform('ADD')
```

```python
# SUBTRACTION of First frame from the second
perform('SUB')
```

```python
perform('MUL')
```

```python
def videocapture(sleep=0.1,arg = None):
    cap = cv2.VideoCapture(0)
    fourcc = cv2.VideoWriter_fourcc(*'mp42') #(*'MP42')
    out = cv2.VideoWriter(arg+'_video.avi', fourcc, 20.0, (640, 480))
    while(True):
        _,frame01 = cap.read()
        time.sleep(sleep)
        _,frame02 = cap.read()
    
        
        if(arg == "ADD"):
            frame = frame01 + frame02
        elif(arg == "SUB"):
            frame = frame02 - frame01
        elif(arg == "MUL"):
            frame = frame01 * frame02
        else:
            pass
        
        frame = cv2.resize(frame, (640, 480))
        out.write(frame)
        
        cv2.imshow('Image',frame)
        if(cv2.waitKey(1) & 0xFF == ord('q')):
            break
    cap.release()
    cv2.destroyAllWindows()
```

```python
videocapture(0.05,'MUL')
```
